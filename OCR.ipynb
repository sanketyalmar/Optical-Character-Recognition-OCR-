```python
from pdf2image import convert_from_path
import easyocr
import numpy as np
import PIL
from PIL import ImageDraw
import spacy
```

```python
reader = easyocr.Reader(['en'])
```

```python
PDF="Sanskrit_Text.pdf"
```

```python
images = convert_from_path('Sanskrit_Text.pdf')
```

```python
from IPython.display import display, Image
display(images[0:4])
```

```python
bounds = reader.readtext(np.array(images[0]), min_size=0, slope_ths=0.2, ycenter_ths=0.7, height_ths=0.6, width_ths=0.8,decoder='beamsearch', beamWidth=10)
bounds
```

```python
def draw_boxes(image, bounds, color='yellow', width=2):
    draw = ImageDraw.Draw(image)
    for bound in bounds:
        p0, p1, p2, p3 = bound[0]
        draw.line([*p0, *p1, *p2, *p3, *p0], fill=color, width=width)
    return image

draw_boxes(images[0], bounds)
```
